```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Image function that allows us to  open up images and be able to transform an image into an array that numpy can understand
from PIL import Image
```

```python
 pic = Image.open('00-puppy.jpg')
```

```python
type(pic)
```

```python
#convert the image into an array
pic_arr = np.asarray(pic)
```

```python
pic_arr.shape
```

```python
#imshow function allows us to show the array as an image
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
pic_red.shape
```

```python
#R G B
#selecting only the green channel
plt.imshow(pic_red[:,:,1])
```

```python
pic_red[:,:,1]
```

```python
#image without the green channel
pic_red[:,:,1] = 0
```

```python
plt.imshow(pic_red)
```
